## Matplotlib example

```python
# import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# prepare data
x = [1, 2, 3, 4]
y = [11, 33, 55, 77]

#setup plot
fig, ax = plt.subplots(figsize=(10, 10))

# plot data
ax.plot(x, y)

# customize plot
ax.set(title="Simple plot",
      xlabel="x-axis",
      ylabel="y-axis")

# save and show
fig.savefig("images/sample-plot.png")
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                            ncols = 2,
                                            figsize = (10, 5))

# Plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows = 2,
                      ncols = 2,
                      figsize = (10, 5))

# Plot each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000));
```

## Pyplot vs. Matplotlib OO method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
# Pyplot method
over_50.plot(kind = "scatter",
            x = "age",
            y = "chol",
            c = "target");
```

```python
# OO method with Pyplot methid
fig, ax = plt.subplots(figsize = (10, 5))
over_50.plot(kind = "scatter",
            x = "age",
            y = "chol",
            c = "target",
            ax = ax);

ax.set_xlim([50,77]);
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize = (10, 6))

# Plot data
scatter = ax.scatter(x = over_50["age"],
                    y = over_50["chol"],
                    c = over_50["target"],);

# Customize the plot
ax.set(title = "Heart disease and Cholesterol Level",
      xlabel = "Age",
      ylabel = "Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title = "Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle = "--");
```

```python
# Style customization
plt.style.use("seaborn-whitegrid")

# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize = (10, 10))

# Add data to ax0
scatter = ax0.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"],
                     cmap = "winter")

# Customize ax0
ax0.set(title = "Heart disease and Cholesterol Level",
      xlabel = "Age",
      ylabel = "Cholesterol")
ax0.set_xlim([50, 78])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title = "Target")

# Add a meanline
ax0.axhline(y = over_50["chol"].mean(),
          linestyle = "--")

# Add data to ax1
scatter = ax1.scatter(x = over_50["age"],
                     y = over_50["thalach"],
                     c = over_50["target"],
                     cmap = "viridis")

# Customize ax1
ax1.set(title = "Heart disease and Max Heart Level",
        xlabel = "Age",
        ylabel = "Max heart rate")
ax1.set_xlim([50, 78])
ax1.set_ylim([60, 200])

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title = "Target")

# Add a meanline
ax1.axhline(y = over_50["thalach"].mean(),
          linestyle = "--");

fig.suptitle("Heart Disease Analysis", fontsize = "15", fontweight = "bold");

fig.savefig("heart-disease-analysis.png");
```
